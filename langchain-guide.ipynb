**Chains**

```python
from langchain.prompts import PromptTemplate
from langchain.llms import OpenAI
from langchain.chains import LLMChain
```

```python
llm = OpenAI(temperature=0.9)
prompt = PromptTemplate(
    input_variables=["animals"],
    template="What are 5 vacation destinations for someone who likes to see {animals}?",
)
```

```python
chain = LLMChain(llm=llm, prompt=prompt)
```

```python
print(chain.run("dolphins"))
```

```python
from langchain.agents import load_tools
from langchain.agents import initialize_agent
from langchain.agents import AgentType
from langchain.llms import OpenAI

```

```python
# First, let's load the language model we're going to use to control the agent.
llm = OpenAI(temperature=0)
```

```python
import os
os.environ["SERPAPI_API_KEY"] = "PVT"
```

```python
# Next, let's load some tools to use. Note that the `llm-math` tool uses an LLM, so we need to pass that in.
tools = load_tools(["serpapi", "llm-math"], llm=llm)
```

```python
agent = initialize_agent(tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, verbose=True)

```

```python
agent.run("What is the average weight of a squirel? What is that number raised to 3?")
```
